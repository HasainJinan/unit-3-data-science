#### 1.5 - Backgrounds

```python
plt.figure(figsize=(10, 6), facecolor='lightgray')
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
plt.title('Gray Figure Background')
plt.show()
```

##### Question 11
What changed when we added `facecolor='lightgray'` to `plt.figure`?
- **Answer:** Adding "facecolor" caused the outer box to change to a light gray color.

##### Question 12
What didnt change in this example? Hyposthesize why the whole graph did not change color.
- **Answer:** The color of the graph background did not change. This is likely because there is another property that will change the graph background instead.

Now let's change the plot area (inner) background

```python
plt.figure(figsize=(10, 6))
ax = plt.gca()  # get current axes
ax.set_facecolor('lightyellow')
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
plt.title('Yellow Plot Area Background')
```

##### Question 13
What's the difference between `plt.figure(facecolor='...')` and `ax.set_facecolor('...')`?
- **Answer:** "plt.figure(facecolor='...')" only changes the color of the margin surrounding the graph and before the outside of the container. "ax.set_facecolor('...')" changes the actual graph's background color. 

##### Question 14
Why have two seperate functions for setting the background? Create a scenario in which that would be useful.
- **Answer:** We likely have two separate functions for setting the background since we might want to change the color of the graph compared to its surrounding to make it stand out more. For example, when adding color to the graph, you may want to have the graph stand out with a bright color against a dark border to make it pop out and be easy to read.

#### 2.0 - Line Plot
Let's start with a simple line plot.

```python
month = [1, 2, 3, 4, 5, 6]
sales = [2, 4, 5, 8, 9, 12]

plt.figure(figsize=(10, 6))
plt.plot(month, sales)
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales (thousands)')
plt.grid(True)
plt.show()
```

#### 2.1 - Scatter Plot
Let's look at some data about students' study time and test scores:

```python
study_hours = [1, 2, 2, 3, 4, 4, 5, 5, 6]
test_scores = [65, 70, 75, 80, 85, 85, 90, 95, 100]

plt.figure(figsize=(10, 6))
plt.scatter(study_hours, test_scores)
plt.title('Study Time vs. Test Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.show()
```

##### Question 18
Compare the above code block to the line plot and examine what is different. Which function was used to create a scatter plot?
- **Answer:** Unlike a line graph, a scatter plot uses two separate, predefined lists and takes each ordered pair in the lists and plots points that way.

#### 2.2 - Bar Plot
Let's look at the favorite pets in a classroom:

```python
categories = ['Dogs', 'Cats', 'Fish', 'Birds']
values = [12, 8, 3, 2]

plt.figure(figsize=(10, 6))
plt.bar(categories, values)
plt.title('Favorite Pets in Class')
plt.xlabel('Pet Type')
plt.ylabel('Number of Students')
plt.show()
```

#### 2.3 - Histogram
Let's look at the distribution of heights in a class:

```python
heights = [65, 68, 67, 67, 70, 65, 68, 69, 66, 71, 
           67, 68, 68, 69, 70, 67, 66, 67, 68, 69]

plt.figure(figsize=(10, 6))
plt.hist(heights, bins=5, edgecolor='black')
plt.title('Student Heights')
plt.xlabel('Height (inches)')
plt.ylabel('Number of Students')
plt.show()
```
